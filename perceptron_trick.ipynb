```python
from sklearn.datasets import make_classification
import numpy as np
x,y=make_classification(n_samples=100, n_features=2, n_classes=2,n_informative=1, random_state=41 , n_clusters_per_class=1, hypercube=False, class_sep=10,n_redundant=0)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap="winter",s=100)
```

```python
def perceptron(x,y):
    x=np.insert(x,0,1,axis=1)
    weights=np.ones(x.shape[1])
    lr=0.1
    for i in range(1000):
        j=np.random.randint(0,x.shape[0])
        y_hat=step(np.dot(weights,x[j]))
        weights=weights+lr*(y[j]-y_hat)*x[j]
    return weights[0], weights[1:]
def step(x):
    return 1 if x>=0 else 0
```

```python
intercept, weights=perceptron(x,y)
print(intercept, weights)
```

```python
m=-weights[0]/weights[1]
c=-intercept/weights[1]
```

```python
x_input=np.linspace(-3,3,200)
y_input=m*x_input+c
plt.figure(figsize=(10,6))
plt.plot(x_input, y_input, color="red",linewidth=2)
plt.scatter(x[:,0],x[:,1],c=y,cmap="winter",s=100)
plt.ylim(-3,3)
```
