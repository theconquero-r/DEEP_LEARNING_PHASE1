```python
import numpy as np

# AND gate dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

target = np.array([0,0,0,1])

# Initialize
w = np.zeros(2)
b = 0
lr = 0.1
epochs = 10

for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for x, d in zip(X, target):

        # Linear output (No threshold)
        y = np.dot(w, x) + b

        # Error
        error = d - y

        # Weight update
        w = w + lr * error * x

        # Bias update
        b = b + lr * error

        print("Input:", x,
              "Target:", d,
              "Output:", round(y,3),
              "Weights:", w,
              "Bias:", round(b,3))

print("\nFinal Weights:", w)
print("Final Bias:", b)
```

```python
import numpy as np

# AND gate dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

target = np.array([0,0,0,1])

# Initialize
w = np.zeros(2)
b = 0
lr = 0.1
epochs = 10

for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for x, d in zip(X, target):

        # Net input
        net = np.dot(w, x) + b

        # Step activation
        if net >= 0:
            y = 1
        else:
            y = 0

        # Error
        error = d - y

        # Weight update
        w = w + lr * error * x

        # Bias update
        b = b + lr * error

        print("Input:", x,
              "Target:", d,
              "Output:", y,
              "Weights:", w,
              "Bias:", b)

print("\nFinal Weights:", w)
print("Final Bias:", b)
```
