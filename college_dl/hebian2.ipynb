```python
import numpy as np

# ============================
# Bipolar Sigmoid Activation
# ============================

def bipolar_sigmoid(x):
    return (2 / (1 + np.exp(-x))) - 1

def bipolar_sigmoid_derivative(y):
    return 0.5 * (1 - y**2)


# ==========================================
# Dataset (Bipolar AND Gate)
# ==========================================

X = np.array([
    [-1,-1],
    [-1, 1],
    [ 1,-1],
    [ 1, 1]
])

target = np.array([-1,-1,-1,1])

lr = 0.1
epochs = 10


# ==========================================
# 1. Hebbian Learning
# ==========================================

def hebbian_learning():

    print("\n========== HEBBIAN LEARNING ==========\n")

    w = np.zeros(2)
    b = 0

    for epoch in range(epochs):

        print(f"\nEpoch {epoch+1}")

        for x,d in zip(X,target):

            net = np.dot(w,x) + b

            y = bipolar_sigmoid(net)

            # Hebbian Update
            w = w + lr*x*y
            b = b + lr*y

            print(f"Input={x} Target={d} Output={round(y,3)}")
            print(f"Weights={np.round(w,3)} Bias={round(b,3)}")

    print("\nFinal Weights :",np.round(w,3))
    print("Final Bias :",round(b,3))


# ==========================================
# 2. Perceptron Learning
# ==========================================

def perceptron_learning():

    print("\n========== PERCEPTRON LEARNING ==========\n")

    w = np.zeros(2)
    b = 0

    for epoch in range(epochs):

        print(f"\nEpoch {epoch+1}")

        for x,d in zip(X,target):

            net = np.dot(w,x)+b

            act = bipolar_sigmoid(net)

            # Bipolar Threshold
            y = 1 if act>=0 else -1

            error = d-y

            w = w + lr*error*x
            b = b + lr*error

            print(f"Input={x} Target={d} Output={y}")
            print(f"Weights={np.round(w,3)} Bias={round(b,3)}")

    print("\nFinal Weights :",np.round(w,3))
    print("Final Bias :",round(b,3))


# ==========================================
# 3. Delta Rule
# ==========================================

def delta_learning():

    print("\n========== DELTA LEARNING ==========\n")

    w = np.zeros(2)
    b = 0

    for epoch in range(epochs):

        print(f"\nEpoch {epoch+1}")

        for x,d in zip(X,target):

            net = np.dot(w,x)+b

            y = bipolar_sigmoid(net)

            error = d-y

            derivative = bipolar_sigmoid_derivative(y)

            delta = error*derivative

            w = w + lr*delta*x
            b = b + lr*delta

            print(f"Input={x} Target={d} Output={round(y,3)}")
            print(f"Weights={np.round(w,3)} Bias={round(b,3)}")

    print("\nFinal Weights :",np.round(w,3))
    print("Final Bias :",round(b,3))


# ==========================================
# Run All
# ==========================================

hebbian_learning()

perceptron_learning()

delta_learning()
```
