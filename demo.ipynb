```python
def fib(n, memo={}):
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fib(n - 1, memo) + fib(n - 2, memo)
    return memo[n]


# Print first 10 Fibonacci numbers
for i in range(10):
    print(fib(i), end=" ")
```
